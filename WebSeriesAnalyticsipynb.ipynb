```python
def Continuos_univariate_Analysis(data, features, figsize=(12, 7), kde=True, bins = None):
    f1, (ax_box1, ax_hist1)= plt.subplots(nrows=2, sharex=True, gridspec_kw={"height_ratios":(0.25, 0.75)}, figsize=figsize)
    sns.color_palette("BuPu_r", as_cmap=True)
    sns.boxplot(data=data, x=features, ax= ax_box1, showmeans=True, color='violet')
    sns.histplot(data=data, x=features, ax=ax_hist1, kde=kde, bins= bins, color="orange") if bins else\
         sns.histplot(data=data, x=features, ax=ax_hist1, kde=kde, color="orange") 
    ax_hist1.axvline(data[features].mean(), color='green', linestyle="--")
    ax_hist1.axvline(data[features].median(), color='orange', linestyle="-")
```

```python
def discrete_univariate_analysis(data, feature,perc=False,n=None):
    total = len(data[feature])
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count+1, 5))
    else:
        plt.figure(figsize=(n+1, 5))
    plt.xticks(rotation = 90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette='flare',
        order =data[feature].value_counts().index.sort_values(ascending = False))
    for p in ax.patches:
        if perc==True:
            label = "{:.1f}%".format(100*p.get_height()/total)
        else:
            label=p.get_height()
        x = p.get_x()+p.get_width()/2
        y = p.get_height()
        ax.annotate(label, (x,y), ha="center", va="center", size=12, xytext=(0,5), textcoords="offset points")
    plt.show()
```

# Model Building - Regression

```python
x=df.drop(['rating'],axis=1)
y=df['rating']
```

```python
x.info()
```

```python
x=pd.get_dummies(x,columns=x.select_dtypes(include=['object','category']).columns.tolist(),drop_first=True)
x.head()
```

```python
x.drop(columns='ongoing',inplace=True)
```

```python
x.info()
```

## Training and Testing data 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
print("no. of sample for train :", x_train.shape[0])
print("no. of sample for test :" ,x_test.shape[0])
```

```python
x_train
```

```python
lin=LinearRegression()
lin.fit(x_train,y_train)
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

We will now use this information to find training and testing data performance

```python
def model_performance(model,predictor,target):
    pred=model.predict(predictor)
    r2=r2_score(target,pred)
    rmse=np.sqrt(mean_squared_error(target,pred))
    
    results=pd.DataFrame({
        "RMSE":rmse,
        "R2 score":r2
    },index=[0])
    
    return results
```

```python
print("Training data performance")
lin_model_train=model_performance(lin,x_train,y_train)
lin_model_train
```

```python
print("Testing data performance")
lin_model_test=model_performance(lin,x_test,y_test)
lin_model_test
```

# Feature selection technique

### 1.sequential search- for small amount of data

### 2.Grid search- it is very expensive search algo

### 3.Baseian search-faster but it overfits

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
reg=LinearRegression()
sfs=SFS(reg,k_features=x_train.shape[1],
       forward=True,floating=False,scoring='r2',n_jobs=-1,cv=5)
sfs=sfs.fit(x_train,y_train)
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
fig1=plot_sfs(sfs.get_metric_dict(),kind='std_err',figsize=(15,5))
plt.xticks(rotation=90)
plt.show()
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
reg=LinearRegression()
sfs=SFS(reg,k_features=35,
       forward=True,floating=False,scoring='r2',n_jobs=-1,cv=5)
sfs=sfs.fit(x_train,y_train)
```
